```python
pressure = pd.read_csv("data/pressure.csv")
wind_speed = pd.read_csv("data/wind_speed.csv")
humidity = pd.read_csv("data/humidity.csv")
wind_direction = pd.read_csv("data/wind_direction.csv")
weather_conditions = pd.read_csv("data/weather_description.csv")
temperature = pd.read_csv("data/temperature.csv")
```

```python
pressure = pressure['Philadelphia']
wind_speed = wind_speed['Philadelphia']
humidity = humidity['Philadelphia']
wind_direction = wind_direction['Philadelphia']
weather_conditions = weather_conditions['Philadelphia']
temperature = temperature['Philadelphia']
```

```python
completed_dataset = pd.DataFrame({'pressure': pressure,'weather_conditions':weather_conditions,'wind_speed':wind_speed,'humidity':humidity,'temperature':temperature,'wind_direction': wind_direction,'temperature':temperature,'wind_direction':wind_direction})
```

```python
completed_dataset = completed_dataset.dropna()
completed_dataset
```

```python
clear_target = completed_dataset[ completed_dataset['weather_conditions'] == 'sky is clear']
```

```python
completed_dataset['weather_conditions'].unique()
```

```python
rain_target_weather = ['light rain',
       'heavy intensity rain', 'moderate rain','thunderstorm with rain',
       'thunderstorm with light rain', 
       'very heavy rain', 'thunderstorm','thunderstorm with heavy rain',
       'light rain and snow',
       'freezing rain',
       'light intensity shower rain', 'shower rain']
```

```python
len(rain_target_weather)
```

```python
rain_target = completed_dataset[ completed_dataset['weather_conditions'] == rain_target_weather[0]]
```

```python
for condition in rain_target_weather[1:]:
    rain_target = pd.concat([rain_target,completed_dataset[ completed_dataset['weather_conditions'] == condition]])
```

```python
clear_target['weather_conditions'] = 0
rain_target['weather_conditions'] = 1
```

```python
clear_target
```

```python
rain_target
```

```python
dataset = pd.concat([clear_target,rain_target])
```

```python
dataset
```

```python
dataset.to_csv('dataset.csv', index=False)
```
